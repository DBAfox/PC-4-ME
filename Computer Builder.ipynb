```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cpu_df = pd.read_csv('data/cpu.csv')
case_df = pd.read_csv('data/case.csv')
cpu_cooler_df = pd.read_csv('data/cpu-cooler.csv')
hdd_df = pd.read_csv('data/internal-hard-drive.csv')
keyboard_df = pd.read_csv('data/keyboard.csv')
mem_df = pd.read_csv('data/memory.csv')
monitor_df = pd.read_csv('data/monitor.csv')
mobo_df = pd.read_csv('data/motherboard.csv')
mouse_df = pd.read_csv('data/mouse.csv')
optical_df = pd.read_csv('data/optical-drive.csv')
os_df = pd.read_csv('data/os.csv')
psu_df = pd.read_csv('data/power-supply.csv')
sound_df = pd.read_csv('data/sound-card.csv')
speakers_df = pd.read_csv('data/speakers.csv')
gpu_df = pd.read_csv('data/video-card.csv')
wifi_df = pd.read_csv('data/wireless-network-card.csv')
```

```python
cpu_df.dropna(subset=['price'], inplace=True)
cpu_df['graphics'].fillna('needs_gpu', inplace=True)
```

```python

# Histogram for 'price'
plt.figure(figsize=(8, 6))
plt.hist(cpu_df['price'], bins=20, color='skyblue')
plt.title('Price Histogram')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Histogram for 'core_count'
plt.figure(figsize=(8, 6))
plt.hist(cpu_df['core_count'], bins=10, color='salmon')
plt.title('Core Count Histogram')
plt.xlabel('Number of Cores')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Histogram for 'core_clock'
plt.figure(figsize=(8, 6))
plt.hist(cpu_df['core_clock'], bins=20, color='lightgreen')
plt.title('Core Clock Histogram')
plt.xlabel('Core Clock (GHz)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Histogram for 'boost_clock'
plt.figure(figsize=(8, 6))
plt.hist(cpu_df['boost_clock'], bins=20, color='orange')
plt.title('Boost Clock Histogram')
plt.xlabel('Boost Clock (GHz)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Histogram for 'tdp'
plt.figure(figsize=(8, 6))
plt.hist(cpu_df['tdp'], bins=20, color='lightblue')
plt.title('TDP Histogram')
plt.xlabel('TDP (W)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Create a figure with subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Scatter plot for 'core_count' vs 'price'
axs[0, 0].scatter(cpu_df['core_count'], cpu_df['price'], color='skyblue')
axs[0, 0].set_title('Core Count vs Price')
axs[0, 0].set_xlabel('Core Count')
axs[0, 0].set_ylabel('Price ($)')

# Scatter plot for 'core_clock' vs 'price'
axs[0, 1].scatter(cpu_df['core_clock'], cpu_df['price'], color='salmon')
axs[0, 1].set_title('Core Clock vs Price')
axs[0, 1].set_xlabel('Core Clock (GHz)')
axs[0, 1].set_ylabel('Price ($)')

# Scatter plot for 'boost_clock' vs 'price'
axs[0, 2].scatter(cpu_df['boost_clock'], cpu_df['price'], color='lightgreen')
axs[0, 2].set_title('Boost Clock vs Price')
axs[0, 2].set_xlabel('Boost Clock (GHz)')
axs[0, 2].set_ylabel('Price ($)')

# Scatter plot for 'tdp' vs 'price'
axs[1, 0].scatter(cpu_df['tdp'], cpu_df['price'], color='orange')
axs[1, 0].set_title('TDP vs Price')
axs[1, 0].set_xlabel('TDP (W)')
axs[1, 0].set_ylabel('Price ($)')

# Scatter plot for 'graphics' vs 'price'
axs[1, 1].scatter(cpu_df['graphics'], cpu_df['price'], color='lightblue')
axs[1, 1].set_title('Graphics vs Price')
axs[1, 1].set_xlabel('Graphics')
axs[1, 1].set_ylabel('Price ($)')

# Rotate the x-axis labels for 'graphics' plot
plt.sca(axs[1, 1])  # Set the current axis to the 'graphics' plot
plt.xticks(rotation=90)  # Rotate the x-axis labels by 90 degrees

# Hide the empty subplot
fig.delaxes(axs[1,2])

plt.tight_layout()
plt.show()
```
